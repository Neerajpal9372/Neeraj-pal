```python
x=sns.countplot(x='Gender' , data=df)
plt.title=('Gender distribution')
for bars in x.containers:
  x.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False)
```

#most of the buyers are female and even purchasing power of female are greater than man


```python
x=sns.countplot(x='Age Group' , data=df , hue='Gender')
plt.title=('Age group distribution ')
for bars in x.containers:
  x.bar_label(bars)
```

#Total amount vs age group

```python
sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending= False)
```

```python
sales_age
```

#most of the buyers are of age group between 26-35yrs male/female

```python
# top 10 states by order
sales_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders' , ascending=False)
```

```python
sales_state
```

```python
sns.barplot(x='State' , y='Orders' , data=sales_state)
plt.show()
```

# Total state by amount


```python
sales_state=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False)
sales_state
```

```python
sales_state=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False)
sns.set(rc={'figure.figsize' : (15,10)})
sns.barplot(x='State' , y='Amount' , data=sales_state)
plt.show()
```

most of the oreders and total sales/amount are form up,maharashtra and karnatka

```python
sns.set(rc={'figure.figsize' : (5,5)})
ax=sns.countplot(x='Marital_Status' , data=df)
for bars in ax.containers:
  ax.bar_label(bars)
plt.show()
```

```python
shadd1=df.groupby(['Marital_Status' , 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False)
shadd1
```

```python
shadd1=df.groupby(['Marital_Status' , 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False)
sns.barplot(data=shadd1, x='Marital_Status' , y='Amount' , hue='Gender')
plt.show()
```

```python
ax=sns.countplot(x='Occupation' , data=df)
sns.set(rc={'figure.figsize' :(25,10)})
for bars in ax.containers:
  ax.bar_label(bars)
plt.show()
```

```python
occ=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False)
sns.barplot(x='Occupation' , y='Amount' , data=occ)
```

```python
ax=sns.countplot(x='Product_Category' , data=df)
sns.set(rc={'figure.figsize' :(25,10)})
for bars in ax.containers:
  ax.bar_label(bars)
plt.show()
```

```python
pr_sales=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False)
sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(x='Product_Category' , y='Amount' , data= pr_sales)
```

most of the sales product are food , cloths and electronics
