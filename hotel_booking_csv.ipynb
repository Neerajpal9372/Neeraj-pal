```python
# Handle missing values
df['children'] = df['children'].fillna(0)
df['country'] = df['country'].fillna('Unknown')
df['agent'] = df['agent'].fillna(0)
df['company'] = df['company'].fillna(0)
```

```python
df.isnull().sum()
```

```python
# Convert date columns to datetime
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])

# Create total guests column
df['total_guests'] = df['adults'] + df['children'] + df['babies']

# Create total nights column
df['total_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
```

```python
# Plot hotel type distribution
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='hotel')
plt.title('Hotel Type Distribution')
plt.show()
```

```python
# Plot cancellation status
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='is_canceled')
plt.title('Booking Cancellation Status')
plt.show()
```

```python
# Plot meal types
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='meal', order=df['meal'].value_counts().index)
plt.title('Meal Type Distribution')
plt.show()
```

```python
# Plot market segments
plt.figure(figsize=(12,5))
sns.countplot(data=df, x='market_segment', order=df['market_segment'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Market Segment Distribution')
plt.show()
```

```python

# Cancellation rate by hotel type
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='hotel', y='is_canceled')
plt.title('Cancellation Rate by Hotel Type')
plt.ylabel('Cancellation Rate')
plt.show()
```

```python
# ADR by hotel type
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='hotel', y='adr')
plt.title('Average Daily Rate by Hotel Type')
plt.show()
```

```python
# Lead time vs cancellation
plt.figure(figsize=(10,5))
sns.boxplot(data=df, x='is_canceled', y='lead_time')
plt.title('Lead Time vs Cancellation Status')
plt.xlabel('Cancelled (1) or Not (0)')
plt.show()
```

```python
# Cancellation rate by market segment
plt.figure(figsize=(12,5))
df.groupby('market_segment')['is_canceled'].mean().sort_values().plot(kind='bar')
plt.title('Cancellation Rate by Market Segment')
plt.ylabel('Cancellation Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
# Monthly bookings
plt.figure(figsize=(12,5))
df['arrival_date_month'].value_counts().plot(kind='bar')
plt.title('Bookings by Month')
plt.show()
```

```python
# Monthly cancellation rate
plt.figure(figsize=(12,5))
df.groupby('arrival_date_month')['is_canceled'].mean().sort_values().plot(kind='bar')
plt.title('Cancellation Rate by Month')
plt.ylabel('Cancellation Rate')
plt.show()
```

```python
# ADR by month
plt.figure(figsize=(12,5))
sns.boxplot(data=df, x='arrival_date_month', y='adr',
            order=['January', 'February', 'March', 'April', 'May', 'June',
                   'July', 'August', 'September', 'October', 'November', 'December'])
plt.xticks(rotation=45)
plt.title('Average Daily Rate by Month')
plt.show()
```
